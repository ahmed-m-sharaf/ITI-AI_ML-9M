### **Stemming**

cutting off the end or the beginning of the word, using `list of common prefixes and suffixes` that can be found in an inflected word.

#### English

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
words = ['cares', 'flies', 'dies', 'mules', 'denied',
           'played', 'died', 'agreed', 'owned', 'humbled', 'sized',
           'meeting', 'stating', 'siezing', 'itemization',
           'sensational', 'traditional', 'reference', 'colonizer',
           'plotted']

stemmed_words = [stemmer.stem(word) for word in words]
print(', '.join(stemmed_words))
```

> **Don't stem the whole sentnece**

```python
samples = 'I am used to playing chess'

# whole sentnece
print(stemmer.stem(samples))
print()

# words
for word in samples.split():
    print(word, ' --> ', stemmer.stem(word))
```

> Snowball stemmer support `different languages`

```python
from nltk.stem.snowball import SnowballStemmer

print(", ".join(SnowballStemmer.languages))
```

#### Arabic

```python
from nltk.stem.snowball import SnowballStemmer

arabic_stemmer = SnowballStemmer("arabic")
arabic_stemmer.stem("الجو سماؤه صافية")
```

```python
from nltk.stem.snowball import ArabicStemmer

ArabicStemmer().stem("الجو سماؤه صافية")
```

```python
samples = 'الجو سماؤه صافية'

# whole sentnece
print(ArabicStemmer().stem(samples))
print()
# words
for word in samples.split():
    print(word, ' --> ', ArabicStemmer().stem(word))
```

### **Lemmatization**
italicized text
same as stemming, but uses `NER Part Of Speech (POS)`

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
words = ['cares', 'flies', 'dies', 'mules', 'denied',
           'played', 'died', 'agreed', 'owned', 'humbled', 'sized',
           'meeting', 'stating', 'siezing', 'itemization',
           'sensational', 'traditional', 'reference', 'colonizer',
           'plotted']

lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print(', '.join(lemmatized_words))
```

```python
samples = 'I am used to playing chess'

# whole sentnece
print(lemmatizer.lemmatize(samples))
print()

# words
for word in samples.split():
    print(word, ' --> ', lemmatizer.lemmatize(word))
```

### **`Stemming` VS `Lemmatization`**

It's `trade-off` <br>
Supposed to be the same and `Lemmatization is more accurate` and keeps the meaning. <br>
**BUT**

```python
print(stemmer.stem("feet"))
print(lemmatizer.lemmatize("feet"))
```

```python
print(stemmer.stem("played"))
print(lemmatizer.lemmatize("played"))
```
