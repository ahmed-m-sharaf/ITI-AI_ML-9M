### Load Embeddings

```python
import gensim
from gensim.models import word2vec
from gensim.models import KeyedVectors

import pandas as pd
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances

cm = sns.light_palette("blue", as_cmap=True)
```

```python
EMBEDDING_FILE = '/content/GoogleNews-vectors-negative300.bin.gz'
word_vectors   = KeyedVectors.load_word2vec_format(EMBEDDING_FILE, binary=True)
```

### Apply on Embeddings
`man - woman = king - queen`

```python
# let's calculate the most similar word to (king - man + woman)
word_vectors.most_similar(positive=['woman', 'king'], negative=['man'], topn=3)
```

```python
word_vectors.most_similar(positive=['woman', 'prince'], negative=['man'], topn=1)
```
